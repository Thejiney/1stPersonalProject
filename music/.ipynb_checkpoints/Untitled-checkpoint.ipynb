```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
filepath='D:/ai/Downloads/Data/music/'
```

```python
df_my=pd.read_csv(f'{filepath}My YouTube Library.csv')
df_my
```

```python
df_my=df_my.drop(df_my[df_my['Track name'].str.split(' - ').str.len()!=2].index,axis=0)
df_my=df_my.drop(['Playlist name','Type','ISRC'],axis=1)
```

```python
df_my
```

```python
df_my['Artist name']=df_my['Track name'].str.split(' - ', expand=True)[0]
df_my['Song name']=df_my['Track name'].str.split(' - ', expand=True)[1]
df_my=df_my.reset_index()
```

```python
dv = webdriver.Chrome()
dv.get('https://songbpm.com/')
```

```python
KEY_lst = []
DURATION_lst = []
BPM_lst = []
for i in range(20):
    elem = dv.find_element(By.NAME, 'query')
    elem.clear()
    elem.send_keys(df_my['Track name'][i])
    btn_elem = dv.find_element(By.ID, 'search-button')
    btn_elem.click()
    time.sleep(10)
    elem2 = dv.find_elements(By.CSS_SELECTOR, 'div.items-center>span')
    KEY_lst.append(elem2[1].text)
    DURATION_lst.append(elem2[3].text)
    BPM_lst.append(elem2[5].text)
    print([i],'번째read')
```

```python
KEY_lst.append(' ')
DURATION_lst.append(' ')
BPM_lst.append(' ')
```

```python
for i in range(21,len(df_my)):
    elem = dv.find_element(By.NAME, 'query')
    elem.clear()
    elem.send_keys(df_my['Track name'][i])
    btn_elem = dv.find_element(By.ID, 'search-button')
    btn_elem.click()
    time.sleep(10)
    elem2 = dv.find_elements(By.CSS_SELECTOR, 'div.items-center>span')
    KEY_lst.append(elem2[1].text)
    DURATION_lst.append(elem2[3].text)
    BPM_lst.append(elem2[5].text)
    print([i],'번째read')
```

```python
KEY_lst
```

```python
df_my['Key']=KEY_lst
df_my['Duration']=DURATION_lst
df_my['BPM']=BPM_lst
df_my
```

```python
df_my.to_csv(f'{filepath}My YouTube Library_mod.csv')
```
